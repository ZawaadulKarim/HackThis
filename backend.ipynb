```python
def find_keywords(filename):
    in_string = textract.process(filename)
    in_string = str(in_string, "utf-8")
    temp = re.sub("[^a-zA-Z]", " ", in_string)
    temp = temp.lower()
    temp = temp.split()
    english_stopwords = stopwords.words("english")
    porter_stemmer = PorterStemmer()
    out = []
    for word in temp:
        if word not in set(english_stopwords) and len(word) > 2:
            out.append(porter_stemmer.stem(word))
    return list(dict.fromkeys(out))
```

```python
list_of_words = find_keywords("calcSyllabus.pdf")
print(list_of_words)
```

```python
final_list = [YoutubeSearch(word, max_results=1).to_dict() for word in list_of_words]
video_ids = [dic['id'] for inner_list in final_list for dic in inner_list]
```

```python
videos_caption_list = {}

for video in tqdm(video_ids):
    try:
        individual_caption_list = YouTubeTranscriptApi.get_transcript(video)
    except:
        continue
    videos_caption_list[video] = individual_caption_list
```

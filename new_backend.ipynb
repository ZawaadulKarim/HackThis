```python
import textract
import nltk
import re
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
nltk.download("stopwords")
from youtube_search import YoutubeSearch
from youtubesearchpython import SearchVideos
from youtube_transcript_api import YouTubeTranscriptApi
from operator import itemgetter
from tqdm import tqdm
import requests
```

```python
from rake_nltk import Rake, Metric
in_string = textract.process('calcSyllabus.pdf')
in_string = str(in_string, "utf-8")
r = Rake(ranking_metric=Metric.WORD_DEGREE)
r.extract_keywords_from_text(in_string)
key_words = r.get_ranked_phrases()
print(key_words)
```

```python
lower_key_words = key_words[0:151]
print(len(lower_key_words))
```

```python
import concurrent.futures
from time import time
```

```python
final_list = []
start = time()
with concurrent.futures.ThreadPoolExecutor(max_workers=15) as executor:
    future_to_result = {executor.submit(YoutubeSearch, word, max_results=1): word for word in lower_key_words}
    for future in concurrent.futures.as_completed(future_to_result):
        result = future_to_result[future]
        try:
            final_list.append(future.result().to_dict())
        except Exception as exc:
            print('generated an exception:')
video_ids = [dic['id'] for inner_list in final_list for dic in inner_list]
print(f'Time taken: {time() - start}')
```

```python
print(video_ids)
```

```python
duration = [dic['duration'] for inner_list in final_list for dic in inner_list]
print((duration))
```

```python
def to_seconds(timestr):
    seconds= 0
    for part in timestr.split(':'):
        seconds= seconds*60 + int(part)
    return seconds
```

```python
for i in range(len(duration)):
    duration[i] = to_seconds(duration[i])
print(duration)

```

```python
def get_caption(video_id):
    return video_id,YouTubeTranscriptApi.get_transcript(video_id)
```

```python
videos_caption_list = {}
start = time()
with concurrent.futures.ThreadPoolExecutor(max_workers=15) as executor:
    future_to_caption = {executor.submit(get_caption,video): video for video in video_ids}
    for future in concurrent.futures.as_completed(future_to_caption):
        caption = future_to_caption[future]
        try:
            res = future.result()
            videos_caption_list[res[0]] = res[1]
        except Exception as exc:
            continue
print(videos_caption_list)
print(f'Time taken: {time() - start}')
```

```python
caption_text_list = {}
for video,data in videos_caption_list.items():
    text = ''
    for caption in data:
        text += caption['text']
    caption_text_list[video] = text
print(caption_text_list)
```

```python
#output is video id:number of matches
num_matches_list = {}
for video,text, seconds in zip (caption_text_list.keys(), caption_text_list.values(), duration):
    
    num_matches = 0
    for word in key_words:
        num_matches+= text.count(word)
    num_matches
    num_matches_list[video]= (num_matches) / seconds
    
    num_matches_list = dict(sorted(num_matches_list.items(),key = lambda kv:(kv[1], kv[0]),reverse = True))
print(num_matches_list)
```

```python
output_list = list(num_matches_list.keys())
```

```python
print(output_list)
```

```python
def final_url_list_generator(user_input_number):
    final_list = []
    new_list = output_list[:user_input_number]
    for item in new_list:
        final_list.append("https://www.youtube.com/watch?v=" + item)
    return final_list
```
